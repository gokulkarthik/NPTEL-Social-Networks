# Lec110 : Link Analysis (Continued) - Convergence in Repeated Matrix Multiplication (Pre-Requisite 1 )


```python
import numpy as np

A = np.mat('1 2; 3 4')
v = np.mat('1; 1')

print A*v
```

```python
import numpy as np

A = np.mat('1 2; 3 4')
v = np.mat('1; 1')
print v
print "#"*20
for i in range(10):
    z = A*v
    z = z/np.linalg.norm(z)#denominator sqrt(a^2+ b^2)
    v = z
    print z
    print "*"*20
```

```python
import numpy as np

A = np.mat('1 2; 3 4')
v = np.mat('4; 11')
print v
print "#"*20
for i in range(10):
    z = A*v
    z = z/np.linalg.norm(z)#denominator sqrt(a^2+ b^2)
    v = z
    print z
    print "*"*20
```

* Inference : No matter where we start from(v), A converges to a constant vector
